```python
import re

# Define a list of attribute names
attributes = ['Location', 'Electricity', 'Internet']

# Using regular expressions to match the attributes in each cell of the column also not case-sensitive
for attribute in attributes:
    df[attribute] = df['Study Facility'].str.extract(rf'({attribute})', flags=re.IGNORECASE)
```

```python
# Question 3.a

import matplotlib.pyplot as plt

# count the number of "Yes" and "No" answers
yes_count = (df['Online Learning Impact(Yes/no)'] == 'Yes').sum()
no_count = (df['Online Learning Impact(Yes/no)'] == 'No').sum()

# plot the chart
fig1, ax = plt.subplots()

bar1 = ax.bar(['Yes'], [yes_count], color='blue')
bar2 = ax.bar(['No'], [no_count], color='red')

plt.xlabel('Answer')
plt.ylabel('Count')
plt.title('Impact of Online Learning on Studying')

# add dotted lines
plt.axvline(0.5, color='gray', linestyle='--')

# add count labels
ax.text(0.02, yes_count+0.5, str(yes_count), color='blue')
ax.text(1.02, no_count+0.5, str(no_count), color='red')


```

```python
#Based on the given data, 977 out of 1634 surveyed students
#responded with "yes," while 657 responded with "no."
#Based on the data provided, it can be concluded that 977 students out of 1634
#surveyed students had some impact from online learning on their ability to study.
#However, the data does not provide information on whether the impact was positive or negative.
#Therefore, it's difficult to determine the overall effectiveness of online learning based on this data alone.
#Further research and analysis, such as asking more specific questions about the nature of the impact,
#may be necessary to fully understand the impact of online learning on students' ability to study.
#end 3.a
```

```python
#3.b
# group the data by faculty and answer
grouped = df.groupby(['Faculty', 'Online Learning Impact(Yes/no)']).size().reset_index(name='Count')

# pivot the data to get a separate column for yes and no counts for each faculty
pivoted = grouped.pivot(index='Faculty', columns='Online Learning Impact(Yes/no)', values='Count').reset_index()

# plot a bar chart for each faculty
fig, ax = plt.subplots(figsize = (10, 7))
pivoted.plot.bar(x='Faculty', y=['Yes', 'No'], ax=ax)
for i, v in enumerate(pivoted['Yes']):
    if np.isfinite(v):
        plt.text(i-0.15, v+0.5, str(v), color='black', fontweight='bold')
for i, v in enumerate(pivoted['No']):
    if np.isfinite(v):
        plt.text(i+0.05, v+0.5, str(v), color='black', fontweight='bold')

plt.xlabel('Faculty')
plt.ylabel('Count')
plt.title('Comparison Bar Chart')
plt.xticks(rotation=0)

plt.show()
```

```python
# It can be concluded that more students in each faculty have had an impact on online learning.
#end of 3.b
```
